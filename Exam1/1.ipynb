## Answer 1.a

Lagrangian is: 
$$L(\omega,\lambda,\gamma)=\frac{1}{2}\omega'\sum\omega + \lambda(m-\omega'\mu)+\gamma(1-\omega'\mathbf{1})$$

First Order Conditions:
$$
\begin{aligned}
\frac{\partial L}{\partial \omega}(\omega,\lambda,\gamma) &= \omega'\sum - \lambda\mu' - \gamma\mathbf{1}' = 0 \\
\\
\frac{\partial L}{\partial \lambda}(\omega,\lambda,\gamma) &= m - \omega'\mu = 0 \\
\\
\frac{\partial L}{\partial \gamma}(\omega,\lambda,\gamma) &= 1 - \omega'\mathbf{1} = 0
\end{aligned}
$$

## Answer 1.b

Define the vector of returns µ as:
$$
\mu=\begin{pmatrix}
0.02 \\
0.07 \\
0.15 \\
0.20
\end{pmatrix}
$$

```python
import numpy as np
mu = np.array([[0.02],
               [0.07],
               [0.15],
               [0.20]])
```

Define the standard deviation matrix S as:

$$
S=\begin{pmatrix}
0.05 & 0 & 0 & 0 \\
0 & 0.12 & 0 & 0 \\
0 & 0 & 0.17 & 0 \\
0 & 0 & 0 & 0.25 \\
\end{pmatrix}
$$

```python
S = np.array([[0.05, 0, 0, 0],
              [0, 0.12, 0, 0],
              [0, 0, 0.17, 0],
              [0, 0, 0, 0.25]])
```

Define the relation matrix R as:

```python
R = np.array([[1, 0.3, 0.3, 0.3],
              [0.3, 1, 0.6, 0.6],
              [0.3, 0.6, 1, 0.6],
              [0.3, 0.6, 0.6, 1]])
```

Next, calculate $\sum$

$$\sum=SRS'=SRS=
\begin{pmatrix}
0.0025 & 0.0018 & 0.00255 & 0.00375 \\
0.0018 & 0.0144 & 0.01224 & 0.018 \\
0.00255 & 0.01224 & 0.0289 & 0.0255 \\
0.00375 & 0.018 & 0.0255 & 0.0625 \\
\end{pmatrix}
$$

And

$$\sum\nolimits^{-1}=
\begin{pmatrix}
455.95854922 & -25.90673575 & -18.28710759 & -12.43523316 \\
-25.90673575 & 127.73459988 & -32.38341969 & -22.02072539 \\
-18.28710759 & -32.38341969 &  63.64630582 & -15.54404145 \\
-12.43523316 & -22.02072539 & -15.54404145 &  29.43005181 \\
\end{pmatrix}
$$

```python
SRS = S @ R @ S
SRS
```

```python
SRS_inv = np.linalg.inv(SRS)
SRS_inv
```


Then, calculate the intermediate values A, B, C, D:

$$A = \mathbf{1}'\sum\nolimits^{-1}\mathbf{1} = 423.6150$$

$$B = \mu'\sum\nolimits^{-1}\mathbf{1} = 6.8070$$

$$C = \mu'\sum\nolimits^{-1}\mu = 0.9065$$

$$D = AC - B^2 = 337.6752$$

```python
I = np.array([[1],
              [1],
              [1],
              [1]])

A = I.T @ SRS_inv @ I
print(f'A: {A[0][0]:.4f}')

B = mu.T @ SRS_inv @ I
print(f'B: {B[0][0]:.4f}')

C = mu.T @ SRS_inv @ mu
print(f'C: {C[0][0]:.4f}')

D = A * C - B**2
print(f'D: {D[0][0]:.4f}')
```
