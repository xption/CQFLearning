### Answer (a)

Define the weight vector ω as:

$$
\omega=\begin{pmatrix}
\omega_A \\
\omega_B \\
\omega_C \\
\omega_D
\end{pmatrix}
$$

Define the vector of returns µ as:
$$
\mu=\begin{pmatrix}
0.02 \\
0.07 \\
0.15 \\
0.20
\end{pmatrix}
$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import fsolve
```

```python
mu = np.array([[0.02],
               [0.07],
               [0.15],
               [0.20]])
```

Define the standard deviation matrix S as:

$$
S=\begin{pmatrix}
0.05 & 0 & 0 & 0 \\
0 & 0.12 & 0 & 0 \\
0 & 0 & 0.17 & 0 \\
0 & 0 & 0 & 0.25 \\
\end{pmatrix}
$$

```python
S = np.array([[0.05, 0, 0, 0],
              [0, 0.12, 0, 0],
              [0, 0, 0.17, 0],
              [0, 0, 0, 0.25]])
```

```python
R = np.array([[1, 0.3, 0.3, 0.3],
              [0.3, 1, 0.6, 0.6],
              [0.3, 0.3, 1, 0.6],
              [0.3, 0.6, 0.6, 1]])

SRS = S @ R @ S
SRS
```

```python
SRS_inv = np.linalg.inv(SRS)
# SRS_inv = np.round(np.linalg.inv(SRS), 4)

SRS_inv

# tt = SRS @ SRS_inv
# tt
```

```python
I = np.array([[1],
              [1],
              [1],
              [1]])
I_t = np.transpose(I)

A = I_t @ SRS_inv @ I
print(f'A: {A[0][0]:.4f}')

B = I_t @ SRS_inv @ mu
print(f'B: {B[0][0]:.4f}')

C = mu.T @ SRS_inv @ mu
print(f'C: {C[0][0]:.4f}')

D = A * C - B**2
print(f'D: {D[0][0]:.4f}')
```
