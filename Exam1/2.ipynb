## Answer (a)

Verify the computation of at least two sensitivities of your choice.

$$
S=\begin{pmatrix}
0.3 & 0 & 0 \\
0 & 0.2 & 0 \\
0 & 0 & 0.15 \\
\end{pmatrix}
$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import fsolve

mu = np.array([[0],
               [0],
               [0]])

stdev = np.array([[0.30],
                  [0.20],
                  [0.15]])

w = np.array([[0.5],
              [0.2],
              [0.3]])

relation = np.array([[1.0, 0.8, 0.5],
                     [0.8, 1.0, 0.3],
                     [0.5, 0.3, 1.0]])
```

```python
S = np.diag(stdev.flatten())
S
```

Define 
$$\sum=SRS'=SRS=
\begin{pmatrix}
0.0900	& 0.0480	& 0.0225 \\
0.0480	& 0.0400	& 0.0090 \\
0.0225	& 0.0090	& 0.0225 \\
\end{pmatrix}
$$

```python
SRS = S @ relation @ S
SRS
```


$$\sum\omega=
\begin{pmatrix}
0.0900	& 0.0480	& 0.0225 \\
0.0480	& 0.0400	& 0.0090 \\
0.0225	& 0.0090	& 0.0225 \\
\end{pmatrix}
\begin{pmatrix}

=\begin{pmatrix}

$$

```python
sigma_w = SRS @ w
sigma_w
```

$$
\omega'\sum\omega=
\begin{pmatrix}

\end{pmatrix}'
\begin{pmatrix}
0.0900	& 0.0480	& 0.0225 \\

\end{pmatrix}
\begin{pmatrix}

\end{pmatrix}
=0.043555
$$

So,
$$\sqrt{\omega'\sum\omega}=0.2087$$

```python
w_sigma_w = w.T @ sigma_w
w_sigma_w_sqrt = np.sqrt(w_sigma_w)
w_sigma_w_sqrt
```

$$\frac{\partial VaR(\omega)}{\partial \omega_1}=\mu_1+z_\alpha * \frac{(\sum\omega)_1}{\sqrt{\omega'\sum\omega}}=-0.6838$$


$$\frac{\partial VaR(\omega)}{\partial \omega_2}=\mu_2+z_\alpha * \frac{(\sum\omega)_2}{\sqrt{\omega'\sum\omega}}=-0.3867$$


$$\frac{\partial VaR(\omega)}{\partial \omega_3}=\mu_3+z_\alpha * \frac{(\sum\omega)_3}{\sqrt{\omega'\sum\omega}}=-0.2207$$

```python
sensitivity = mu + (-2.326) * sigma_w / w_sigma_w_sqrt
sensitivity
print("Sensitivity of the portfolio at 1% VaR level: ")
print(sensitivity)

sensitivity_1 = sensitivity.flatten()[0]
sensitivity_2 = sensitivity.flatten()[1]
sensitivity_3 = sensitivity.flatten()[2]

print(f"Sensitivity of asset 1: {sensitivity_1:.4f}")
print(f"Sensitivity of asset 2: {sensitivity_2:.4f}")
print(f"Sensitivity of asset 3: {sensitivity_3:.4f}")
```
